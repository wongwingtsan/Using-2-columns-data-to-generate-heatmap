```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set_theme()
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./speed_station_dataset_test/017-0143.csv',skiprows=0,na_values='missing')
```

```python
df
```

```python
selected_columns=df[["Time","Volume","AvgSpeed"]]
```

```python
DF=selected_columns.copy()
```

```python
DF
```

```python
DF=DF.set_index('Time')
```

```python
DF1=DF.drop(columns=['Volume'])
DF1
```

```python
DF2=DF.drop(columns=['AvgSpeed'])
DF2
```

```python
fig, (ax,ax2) = plt.subplots(ncols=2)
fig.subplots_adjust(wspace=0.01)
sns.heatmap(DF1, cmap="Greys", ax=ax, cbar=False)
fig.colorbar(ax.collections[0], ax=ax,location="left", use_gridspec=False, pad=0.2)
sns.heatmap(DF2, cmap="Greys", ax=ax2, cbar=False)
fig.colorbar(ax2.collections[0], ax=ax2,location="right", use_gridspec=False, pad=0.2)
ax2.yaxis.tick_right()
ax2.tick_params(rotation=0)
plt.show()
```
